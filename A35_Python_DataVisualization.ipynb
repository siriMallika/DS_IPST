# <font color='red'>01</font>  นำเข้าชุดข้อมูล  

```python
import pandas as pd
```

```python
#อ่านไฟล์ มาเก็บใน Dataframe
df=pd.read_csv("/content/testscoreNew_c.csv")
df.head(10)
```

```python
#ขอดูข้อมูลแบบมีเงื่อนไข นักเรียนที่มีคะแนนคณิตศาสตร์มากกว่า 50 คะแนน  
df[df['math']>50] #df[df.math>50]
```

```python
#เลือกข้อมูลเฉพาะ math thai science english และ program
df=df[['math','thai','science','english','program']] #'math','thai','science','english','program'
```

```python
df.head(10)
```

# <font color='red'>02</font>  การทำข้อมูลให้เป็นภาพ Data Visualization

```python
#เรียกใช้โมดูลในการสร้างกราฟ  matplotlib  และ  seaborn
import matplotlib.pyplot as plt
import seaborn as sbn
```

## เปรียบเทียบจำนวนนักเรียนที่เข้าสอบแต่ละห้อง

```python
#kind = bar,barh,pie
df.program.value_counts().plot(kind='bar')
```

```python
#kind = bar,barh,pie
df.program.value_counts().plot(kind='barh')
```

```python
#kind = bar,barh,pie
df.program.value_counts().plot(kind='pie')
```

## เปรียบเทียบคะแนนของแต่ละวิชาของนักเรียนแต่ละห้อง

```python
pvt=df.pivot_table(index=['program'])
pvt.plot(kind='bar')
```

```python
pvt=df.pivot_table(index=['program'])
pvt.plot(kind='barh')
```

```python
pvt=df.pivot_table(index=['program'])
pvt.plot(kind='pie',subplots=True)
```

## เปรียบเทียบคะแนนเฉลี่ยแต่ละวิชาของห้อง EP

```python
pvt=df[df.program=='EP'].pivot_table(index=['program']) #EP
pvt.plot(kind='barh')
```

```python
#อยากรู้ว่ามีกี่คนในแต่ละแผนการเรียนที่ได้คะแนนวิชาคณิตศาสตร์มากกว่า 40 คะแนน
pvt=df[df.math>40].pivot_table(index=['program'],values=['math'],aggfunc=['count']) #>40 #program #math
pvt.plot(kind='barh').grid() #bar,barh
```

## เปรียบเทียบการกระจายข้อมูลของแต่ละวิชา

```python
sbn.boxplot(data=df[['math','thai','science','english']]) #'math','thai','science','english'
```

## ดูความสัมพันธ์และการกระจายของข้อมูล

```python
sbn.scatterplot(x='math',y='science',data=df) #'math','science'
```
